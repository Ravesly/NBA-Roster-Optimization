```python
import pandas as pd
import numpy as np
df= pd.read_csv('csv/officials.csv')
df2= pd.read_csv('csv/cleaned/inactive_players_cleaned.csv')
print(df.columns)
print(df.shape)
```

```python
#SEGMENTAR GRUPO DE ESTUDIO
df = df[df['game_id'].isin(df2['game_id'])]
# Verificamos cu√°ntas filas quedaron
print(f"Registros en el grupo de estudio: {len(df)}")
print(df.columns)
```

```python
df.info
```

```python
# Quitamos los decimales para que se vea como un uniforme real y rellenamos los nulos con -1
df['jersey_num'] = df['jersey_num'].fillna(-1).astype(int)

# Creamos una columna de nombre completo 
df['full_name'] = df['first_name'].astype(str) + " " + df['last_name'].astype(str)

#Reordenamos o eliminamos columnas si es necesario para que sea ligero
df_officials_final = df[['game_id', 'official_id', 'full_name', 'jersey_num']].copy()

# Verificamos que no haya duplicados de oficial por partido
df_officials_final.drop_duplicates(subset=['game_id', 'official_id'], inplace=True)
```

```python
#REVISION DE NULOS
df.isnull().sum()
```

```python
print(f"Registros en el grupo de estudio: {len(df_officials_final)}")
df_officials_final.to_csv('csv/cleaned/officials_cleaned_final.csv', index=False)
```
