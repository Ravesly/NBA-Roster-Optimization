```python
import pandas as pd
import datetime
df = pd.read_csv('csv/line_score.csv')
print('ESTE ES EL SHAPE:')
print(df.shape)
print('INFO DE DF')
df.info
```

```python
#CORREGIR FECHAS
df['game_date_est'] = pd.to_datetime(df['game_date_est'])
#SEGMENTAR GRUPO DE ESTUDIO
df = df[df['game_date_est'] >= '2004-01-01']
print(df.columns)
```

```python
#NUEVO SHAPE
print(df.shape)
#REVISION DE NULOS
df.isnull().sum()

#El 46% de las filas de las columnas pts_ot5_home a pts_ot10_home, son nulas
```

```python
#VERIFICAMOS QUE NO HAYA ERRORES DE REGISTROS EN LAS COLUMNAS pts_ot
# Usamos filter para agarrar todas las que tengan 'pts_qtr' o 'pts_ot'
cols_puntos_home = [c for c in df.columns if ('pts_qtr' in c or 'pts_ot' in c) and '_home' in c]
cols_puntos_away = [c for c in df.columns if ('pts_qtr' in c or 'pts_ot' in c) and '_away' in c]

# Creamos una columna temporal con la suma calculada
df['suma_calculada_home'] = df[cols_puntos_home].sum(axis=1)
df['suma_calculada_away'] = df[cols_puntos_away].sum(axis=1)

# Comparamos la suma calculada contra la columna total del dataset (pts_home y pts_away)
# Usamos un margen de error mínimo por si hay nulos (fillna(0))
error_home = df[df['suma_calculada_home'] != df['pts_home'].fillna(0)]
error_away = df[df['suma_calculada_away'] != df['pts_away'].fillna(0)]

# Resultados
print(f"Filas con errores en suma de Local: {len(error_home)}")
print(f"Filas con errores en suma de Visitante: {len(error_away)}")

if len(error_home) == 0 and len(error_away) == 0:
    print("¡Perfecto! Todos los puntajes coinciden con la suma de sus cuartos.")
else:
    print("Se encontraron discrepancias. Revisa las variables 'error_home' o 'error_away'.")
```

```python
#Como encontramos una sola fila de errores, forzamos la suma
df.loc[df['suma_calculada_away'] != df['pts_away'], 'pts_away'] = df['suma_calculada_away']
```

```python
#QUITAMOS LAS COLUMNAS TEMPORALES
df.drop(columns=['suma_calculada_home', 'suma_calculada_away'], inplace=True)
```

```python
print(df.columns)
```

```python
# Filtramos las columnas que van de pts_ot5_away hasta pts_ot10_away
columnas_ot_lejanas1 = [f'pts_ot{i}_away' for i in range(5, 11)]
columnas_ot_lejanas2 = [f'pts_ot{i}_home' for i in range(5, 11)]

# Mostramos cuántos nulos hay en cada una
print("Nulos en tiempos extra lejanos (Visitante):")
print(df[columnas_ot_lejanas1].isnull().sum())
print(df[columnas_ot_lejanas2].isnull().sum())
```

```python
# Contamos cuántos partidos tuvieron puntos en el OT 5
partidos_ot5 = df[df['pts_ot5_away'] > 0]

print(f"Total de partidos desde 2004: {len(df)}")
print(f"Partidos que llegaron a 5 tiempos extra: {len(partidos_ot5)}")
```

```python
# Verificamos los valores únicos en OT5 para entender qué está pasando
print("Valores únicos encontrados en pts_ot5_home en adelante:")

for i in columnas_ot_lejanas1:
    print(df[i].unique())
    # Vemos las primeras filas donde esa columna NO es cero ni nula
    muestra_datos = df[df[i] > 0][['game_id', 'game_date_est', i, 'pts_home']]
    print("\nMuestra de partidos con datos en OT5 en adelante:")
    display(muestra_datos.head())
```

```python
# Verificamos los valores únicos en OT5 para entender qué está pasando
print("Valores únicos encontrados en pts_ot5_away en adelante:")

for i in columnas_ot_lejanas2:
    print(df[i].unique())
    # Vemos las primeras filas donde esa columna NO es cero ni nula
    muestra_datos = df[df[i] > 0][['game_id', 'game_date_est', i, 'pts_away']]
    print("\nMuestra de partidos con datos en OT5 en adelante:")
    display(muestra_datos.head())
```

```python
#COMO NO EXISTEN JUEGOS QUE TENGAN 5TO TIEMPO EN ADELANTE, ELIMINAMOS LAS COLUMNAS pts_ot5 en adelante
for i in columnas_ot_lejanas1:
    df.drop(columns=[i], inplace=True)
for i in columnas_ot_lejanas2:
    df.drop(columns=[i], inplace=True)
```
