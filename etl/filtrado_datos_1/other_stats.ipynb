```python
import pandas as pd
import numpy as np
df= pd.read_csv('csv/other_stats.csv')
df2= pd.read_csv('csv/cleaned/inactive_players_cleaned.csv')
print(df.columns)
print(df.shape)
```

```python
#SEGMENTAR GRUPO DE ESTUDIO
df = df[df['game_id'].isin(df2['game_id'])]
df.info
```

```python
#REVISION DE NULOS
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
#Convertimos las métricas a enteros (int) para eliminar los decimales .0
#Seleccionamos las columnas que queremos limpiar visualmente
cols_a_limpiar = [
    'pts_paint_home', 'pts_paint_away', 
    'pts_2nd_chance_home', 'pts_2nd_chance_away',
    'pts_fb_home', 'pts_fb_away',
    'pts_off_to_home', 'pts_off_to_away',
    'team_rebounds_home', 'team_rebounds_away',
    'team_turnovers_home', 'team_turnovers_away'
]
# Aplicamos la conversión solo a esas columnas
df[cols_a_limpiar] = df[cols_a_limpiar].astype(int)
```

```python
#Verificación final de nulos
print("--- Verificación Final de Nulos en Other Stats ---")
print(df.isnull().sum().sum()) # Debería dar 0

display(df.head())
```

```python
df.to_csv('csv/cleaned/other_stats_cleaned_final.csv', index=False)
```
