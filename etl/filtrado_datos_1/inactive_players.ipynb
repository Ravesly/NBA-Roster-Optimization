```python
import pandas as pd
import numpy as np
df = pd.read_csv('csv/inactive_players.csv')
df2= pd.read_csv('csv/cleaned/line_score_cleaned.csv')
df3= pd.read_csv('csv/cleaned/game_summary_cleaned.csv')
print('ESTE ES EL SHAPE:')
print(df.shape)
print(df.columns)
print('INFO DE DF')
df.info
```

```python
#SEGMENTAR GRUPO DE ESTUDIO
df = df[df['game_id'].isin(df2['game_id'])]
# Verificamos cuántas filas quedaron
print(f"Registros en el grupo de estudio: {len(df)}")
print(df.columns)
```

```python
print(df.shape)
df.info
```

```python
# Eliminamos si el mismo jugador aparece dos veces en el mismo partido
df.drop_duplicates(subset=['game_id', 'player_id'], keep='first', inplace=True)
print(df.shape)
```

```python
#Seguimos con la columna Jersey.
jugadores_con_cero = df[df['jersey_num'] == 0]
print(f"Jugadores que realmente usan el número 0: {len(jugadores_con_cero)}")

#Al ser cero un numero valido de camiseta, reemplazamos los nulos por un número imposible (-1)
df['jersey_num'] = df['jersey_num'].fillna(-1).astype(int)

#Convertimos la columna a enteros
df['jersey_num'] = df['jersey_num'].astype(int)
```

```python
#REVISION DE NULOS
df.isnull().sum()

#REVISION VALORES UNICOS COLUMNAS team_city y  team_name
print(df['team_city'].unique())
print(df['team_name'].unique())
```

```python
# Verificamos los registros erroneos y los pertenecientes a AllStars

matriz_equipos= ['Team','East NBA All Stars', 'West NBA All Stars', 'Real Madrid']
game_id=[]
for i in matriz_equipos:
    df_temporal= df[df['team_city'] == i]
    if i == 'Team':
        print(df_temporal['team_name'].value_counts())
    game_id.append(df_temporal['game_id'])
    #guardamos todos los id en game_id
    print(df_temporal['team_name'].unique())
    print(df_temporal.shape)
```

```python
#Eliminamos filas pertenecientes a los registros de AllStars
# 1. Usamos concatenate para unir todas las sub-listas en una sola
lista_negra_limpia = np.concatenate(game_id).astype(str)

# 2. Aplicamos el filtro de negación (~) asegurando que el DF también sea string
df = df[~df['game_id'].astype(str).isin(lista_negra_limpia)]

# 3. Verificamos la limpieza
print(f"Registros inactive_players después de la depuración: {len(df)}")

# 4. Limpiamos line_Score_cleaned (df2)
print(f"Registros line_Score_cleaned antes de la depuración: {len(df2)}")
df2 = df2[~df2['game_id'].astype(str).isin(lista_negra_limpia)]

# 5. Limpiamos Game_summary_cleaned (df3)
print(f"Registros Game_summary_cleaned antes de la depuración: {len(df3)}")
df3 = df3[~df3['game_id'].astype(str).isin(lista_negra_limpia)]

# 4. Verificación de resultados
print("--- Auditoría de Limpieza Final ---")
print(f"Partidos en Line Score (df2): {len(df2)}")
print(f"Partidos en Game Summary (df3): {len(df3)}")
```

```python
df.to_csv('csv/cleaned/inactive_players_cleaned.csv', index=False)
```
